## Two ways to use matplotlib

#### 1. Object-oriented style (OO-style)

```python
np.linspace?
```

```python
np.linspace(0, 2, 4)
```

```python
x = np.linspace(0, 2, 100)

# note that even in oo-style we have to use plt.figure

# create figure and axes
fig, ax = plt.subplots()

# plot some data on the axes
ax.plot(x, x, label='Linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

# set an x and y label to the axes
ax.set_xlabel('x label')
ax.set_ylabel('y label')

# set the tile 
ax.set_title('OO-STYLE PLOT')
plt.legend(loc = 'upper left')
plt.grid()
plt.axhline()
plt.axvline()
plt.show()
```

#### 2. pyplot-style 

```python
# plot some data on the implicit axes
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# set the x and y label
plt.xlabel('x label')
plt.ylabel('y lable')

# set the title
plt.title('pyplot style')
plt.legend()
# plt.legend(loc = 'upper left')
plt.grid()
plt.axhline()
plt.axvline()
plt.show()
```

## Simple plot

```python
# Generate two random numbers
x = np.linspace(1, 10, 20)
y = x * np.linspace(1, 10, 20)

# create basic plot
plt.plot(x, y)

# show the plot
plt.show()
```

## Adding element to  the plot

To understand the plot better we will add the element like label, and limits

```python
plt.plot(x, y)
plt.title('Basic plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(0,12)
plt.ylim(0,120)
plt.show()
```

## Add colour, markers, line customization

```python
plt.plot(x, y, c='b', linestyle='-', lw=5)
plt.plot(x, y, c='w', linestyle='--', linewidth=2)
plt.show()
```

#### Marker and Marker size and Marker color

```python
plt.plot(x, y, c='r', linestyle='-', linewidth=2, marker='*', ms=20, mfc='g', mec = 'orange')
plt.show()
```

#### Grid()

```python
plt.plot(x, y)
# plt.grid(True, c='g', ls='-', lw=3)
plt.grid(True, c='r', ls='--', lw=1)
```

```python
# grid on either x or y axis
plt.plot(x, y)
plt.grid(True, c='g', axis='y')
plt.grid(True, c='r', axis='x', ls='--', lw=1)
```

## Set Tick and Tick Labels

```python
plt.plot(x, x**3)
plt.yticks([100,400, 600, 800, 1000])
plt.xticks([0,2,4,6,8,10, 12])
plt.show()
```

#### Single subplot

```python
# create subplot of 2 rows and 2 columns
fig, ax = plt.subplots(2, 2)

# increase the figure size
# plt.rcParams['figure.figsize'] = [12, 8]
plt.figure(figsize=(20, 16))

ax[0, 0].plot(x**-0.2, x)
ax[0, 0].set_title('First subplot')

ax[0, 1].plot(x**2.5, x)
ax[0, 1].set_title('Second subplot')

ax[1, 0].plot(x**12, x)
ax[1, 0].set_title('Third subplot')

ax[1, 1].plot(x**3, x)
ax[1, 1].set_title('Fourth subplot')

# give proper spacing between the subplot
plt.tight_layout()
plt.show()
```

## Axes Object

```python
# let's add axes using add_axes() method
# create a sample data

y = [1,5,10,15,20,30]
x1 = [1,10,20,30,45, 55]
x2 = [1,30,45,80,90,122]

# create a figure
fig = plt.figure(figsize=(5, 5))

# add the axes
ax = fig.add_axes([0,0,2,1])
l1 = ax.plot(x1, y, 'sy-', label = 'line 1')
l2 = ax.plot(x2, y, 'gv--', label = 'line 2')
ax.plot(y, y, 'rs--', label = 'line 3')

# add additional paramaters
# ax.legend(labels=('line 1' 'line 2'), loc='lower right')
ax.legend()
ax.set_title('Use of adding axes title')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

## Different types of matplotlib plot

### 1. Bar graph

```python
# x-axis show the subject and y-axis show the marks
x = ['maths', 'science', 'english', 'hindi']
y = [75, 95, 70, 92]

plt.bar(x, y)
plt.show()
```

#### Let's do some customization

```python
plt.bar(x, y, color='g', width=0.5, bottom=10, align='center', edgecolor='r', linewidth=1.4, tick_label=x)
plt.show()
```

```python
[xx + str(yy) for xx, yy in zip(x, y)]
```

```python
plt.bar?
```

```python
plt.bar(x, y, color='y', width=0.2, bottom=50, align='center', edgecolor='b', linewidth=2.8, 
        tick_label=[xx + str(yy) for xx, yy in zip(x, y)])
plt.show()
```

#### horizontal bar

```python
plt.barh(x, y, color = ['y', 'g', 'r', 'b'])
plt.show()
```

### 2. pie chart

```python
# assume that we have a data on the number of ticket resolve in a month
# the manager wants to know inter of the ticket closed in that particular week

ticket_close = [13, 54, 16, 13, 23, 57]
agent = ['arjun', 'ben', 'chetan', 'delip', 'eshan', 'farahan']

plt.pie(ticket_close, labels=agent)
plt.show()
```

#### customization

```python
# to move one of the wedges of the plot
exploded = [0.2,0.1, 0.4, 0.1,0.1,0.1]

# plt.pie(ticket_close, labels=agent, explode=exploded, autopct='%1.1f%%')
plt.pie(ticket_close, labels=agent, explode=exploded, autopct='%1.2f%%')
plt.show()
```
