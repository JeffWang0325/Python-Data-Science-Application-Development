# Classification

```python
y = df['survived']
X = df.drop('survived', axis=1)
X
```

```python
X.isna().sum()
```

```python
891 * 0.2
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
# X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf2 = RandomForestClassifier()
clf2.fit(X_train, y_train)
clf2.score(X_test, y_test)
```

```python
from joblib import dump, load
dump(clf, 'model.joblib')
```

```python
def convert_sex(str1):
    return 1 if str1 == 'male' else 0

def convert_age(str1):
    bins = [0, 15, 25, 35, 60, 100]
    print(type(pd.cut(pd.Series(str1), bins, labels=[0, 4, 3, 2, 1])))
    return pd.cut(pd.Series(str1), bins, labels=[0, 4, 3, 2, 1])[0]

def convert_embark_town(str1):  
    dict1={'Southampton':0, 'Cherbourg':1, 'Queenstown':2}
    return dict1[str1]

# pclass	sex	age	sibsp	parch	fare	adult_male	embark_town
X1 = [[2, convert_sex('male'), convert_age(50), 1, 0, 20, 1, convert_embark_town('Cherbourg')],
      [2, convert_sex('female'), convert_age(20), 1, 0, 71, 1, convert_embark_town('Queenstown')]]

X1 = scaler.transform(X1)
X1
```
